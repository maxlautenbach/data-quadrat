### Import dataset from prepared csv file

```python
import math

import data_preprocessing

train, test = data_preprocessing.preprocess_data("Residual Load Dataset.csv")
```

### Calculate the Median of all train values for the baseline

```python
y_median = train["Residual Load"].median()
y_median
```

### Calculate the RMSE for this approach

```python
import numpy as np

y_pred = np.full((len(test)), y_median)
y_true = test["Residual Load"]
```

```python
from sklearn.metrics import *
import math

math.sqrt(mean_squared_error(y_true=y_true, y_pred= y_pred))
```

### Baseline Plot

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(12,8))
ax.scatter(x=range(len(y_true.index)), y=y_true)
ax.plot(range(len(y_true.index)), y_pred, c="red", linewidth=3)
ax.set_title("Baseline based on median")
ax.set_xlabel("Index of time-points")
ax.set_ylabel("Residual load in kWh")
```
