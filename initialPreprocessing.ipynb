```python
import numpy as np
#All preprocessing imports
import pandas as pd
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
```

```python
energy_consumption_data = pd.read_csv("Energy Consumption Dataset.csv", index_col=0)
energy_consumption_data.head()
```

```python
energy_consumption_data.describe()
```

```python
#Doing split to normalize each data part
train, test = train_test_split(energy_consumption_data, shuffle=False, test_size=0.2)

train.head()
```

```python
columns = list(energy_consumption_data.columns)
```

```python
from tqdm import tqdm

# create all combinations of considered columns
#combinations = itertools.combinations(features, 2)

# create a figure and specify its size
fig = plt.figure(figsize=(15,40))

# go through all combinations and create one plot for each
figure_index = 1
for column in tqdm(columns[1:]):
    # add a sub plot to the figure
    axs = fig.add_subplot(10,2,figure_index)
    axs.hist(train[column], bins=15)
    axs.set_title(column)
    # increase the figure index (otherwise all plots are drawn in the first subplot)
    figure_index+=1

    
# show the plot
plt.show()
```

```python
#Drop Faulty Data

import numpy as np

train = train.replace(-999, np.NaN)
train = train.dropna()
```
