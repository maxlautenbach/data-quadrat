```python
#All preprocessing imports
import pandas as pd
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
```

```python
energy_consumption_data = pd.read_csv("Energy Consumption Dataset.csv")
energy_consumption_data.head()
```

```python
#Doing split to normalize each data part
test, train = train_test_split(energy_consumption_data, shuffle=False, test_size=0.2)

train.head()
```

```python
# create a list of all columns that we are considering
columns = [ 'Wind Velocity for Station 02014', 'Sun Duration for Station 02014', 'Air Temperature for Station 02014', 'Precipitation Amount for Station 02014' ]

# create all combinations of considered columns
#combinations = itertools.combinations(features, 2)

# create a figure and specify its size
fig = plt.figure(figsize=(15,10))

# go through all combinations and create one plot for each
figure_index = 1
for column in columns:
    # add a sub plot to the figure
    axs = fig.add_subplot(2,3,figure_index)
    axs.hist(energy_consumption_data[column])   
    # increase the figure index (otherwise all plots are drawn in the first subplot)
    figure_index+=1

    
# show the plot
plt.show()
```
