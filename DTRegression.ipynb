```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
import data_preprocessing

energy_consumption_data_train, energy_consumption_data_test = data_preprocessing.preprocess_data("Energy Consumption Dataset.csv")
```

```python
energy_consumption_data_train.head()
```

```python
X_train = pd.DataFrame(energy_consumption_data_train.iloc[:, :18])
Y_train = pd.DataFrame(energy_consumption_data_train.iloc[:, -1])

X_test = pd.DataFrame(energy_consumption_data_test.iloc[:, :18])
Y_test = pd.DataFrame(energy_consumption_data_test.iloc[:, -1])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
def do_reg(estimator):
    estimator.fit(X_train, Y_train)

    # predict values
    Y_pred = estimator.predict(X_test)


    # estimate example using (normalized) RMSE
    from numpy import sqrt 

    from sklearn.metrics import mean_squared_error

    rsme = sqrt(mean_squared_error(Y_test, Y_pred))
    _range = Y_test.max() - Y_test.min()

    return Y_pred, rsme / _range
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import KFold
```

```python
def rand_search_dt():

    model = DecisionTreeRegressor()

    param_dist = {"max_depth":          [ None ] + [i for i in range(1, 9)],
                  "max_features":       [ None ] + [i for i in range(1, 9)],
                  "min_samples_leaf":   [ None ] + [i for i in range(1, 9)],
                  "criterion":          ['squared_error', 'poisson', 'friedman_mse', 'absolute_error'],
                  'ccp_alpha':         [ None ] + [i for i in range(1, 9)]}

    # Create the GridSearchCV object
    rand_search_dt = RandomizedSearchCV(estimator=model, 
                                        param_distributions=param_dist,
                                        cv=KFold(n_splits=5, shuffle=True, random_state=1)
                                       )

    # Fit the GridSearchCV object to the training data
    rand_search_dt.fit(X_train, Y_train)

    # Get the best hyperparameters
    best_dt = rand_search_dt.best_estimator_
    best_params_dt = rand_search_dt.best_params_
    
    return best_dt, best_params_dt
```

```python
def rand_search_rf():
    
    from scipy.stats import uniform, truncnorm, randint

    model = RandomForestRegressor()

    random_grid = {'bootstrap': [True, False],
                   'max_depth': randint(1, 100),
                   'max_features': ['auto', 'sqrt'],
                   'min_samples_leaf': [1, 2, 4],
                   'min_samples_split': [2, 5, 10],
                   'n_estimators': randint(4, 200),
                   'ccp_alpha': randint(0, 10)}


    # Create the GridSearchCV object
    rand_search_rf = RandomizedSearchCV(estimator=model,
                                        param_distributions=random_grid,
                                        scoring="neg_mean_squared_error",
                                        cv=KFold(n_splits=5, shuffle=True, random_state=1)
                                       )

    # Fit the GridSearchCV object to the training data
    rand_search_rf.fit(X_train, Y_train)

    # Get the best hyperparameters
    best_rf = rand_search_rf.best_estimator_
    best_params_rf = rand_search_rf.best_params_
    
    return best_rf, best_params_rf
```

## Optimize Hyperparameters using Grid Search

```python
from sklearn.model_selection import GridSearchCV


# Define the hyperparameter grid
param_grid = {
    'n_estimators': [100],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5, 10]
}

# Create the Decision Tree Regressor
model = RandomForestRegressor()

# Create the GridSearchCV object
rand_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error')

# Fit the GridSearchCV object to the training data
rand_search.fit(X_train, Y_train)

# Get the best hyperparameters
best_params = rand_search.best_params_
best_params
```
