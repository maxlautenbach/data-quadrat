```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
import data_preprocessing

energy_consumption_data_train, energy_consumption_data_test = data_preprocessing.preprocess_data("Energy Consumption Dataset.csv")
```

```python
energy_consumption_data_train.head()
```

```python
energy_consumption_data_test.head()
```

```python
X_train = pd.DataFrame(energy_consumption_data_train.iloc[:, :18])
Y_train = pd.DataFrame(energy_consumption_data_train.iloc[:, -1])

X_test = pd.DataFrame(energy_consumption_data_test.iloc[:, :18])
Y_test = pd.DataFrame(energy_consumption_data_test.iloc[:, -1])
_range = (Y_test.max() - Y_test.min()).item()
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
# try default random forest ( no parameter tuning yet )
regressor = RandomForestRegressor()
regressor.fit(X_train, Y_train)

# predict values
Y_pred = regressor.predict(X_test)
Y_test.values[:5].swapaxes(0, 1), Y_pred[:5]
```

```python
# estimate example using RMSE (normalized) 
from sklearn.metrics import mean_squared_error

rsme = np.sqrt(mean_squared_error(Y_test, Y_pred))
rsme_norm = rsme / _range

print(f"{rsme=}")
print(f"{rsme_norm=}")
```

```python
# inspect example estimator (tree)
import math
trees = regressor.estimators_
print(f"num_trees={len(trees)}")

node_counts = [tree.tree_.node_count for tree in trees]
depths = [tree.get_depth() for tree in trees]
leaves_counts = [tree.get_n_leaves() for tree in trees]

avg_nodes = math.fsum(node_counts) / len(trees)
avg_depth = math.fsum(depths) / len(trees)
avg_leaves = math.fsum(leaves_counts) / len(trees)

print(f"{avg_nodes=}")
print(f"{avg_depth=}")
print(f"{avg_leaves=}")
```

```python
# find better hyperparameters using randomized search

from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import KFold
```

```python
from scipy.stats import randint

regressor = RandomForestRegressor()

random_grid = {'bootstrap': [True, False],
               'max_depth': randint(1, 18),
               'max_features': ['auto', 'sqrt'],
               'min_samples_leaf': [2, 4, 6],
               'min_samples_split': [2, 5, 10],
               'n_estimators': randint(4, 200),
               'ccp_alpha': randint(0, 10)}


# Create the GridSearchCV object
rand_search_rf = RandomizedSearchCV(estimator=regressor,
                                    param_distributions=random_grid,
                                    scoring="neg_mean_squared_error",
                                    cv=KFold(n_splits=5, shuffle=True, random_state=1))

# Fit the GridSearchCV object to the training data
rand_search_rf.fit(X_train, Y_train)

# Get the best hyperparameters
best_rf = rand_search_rf.best_estimator_
best_params_rf = rand_search_rf.best_params_
```

```python
# inspect tuned hyperparameters
best_params_rf
```

```python
# try tuned random forest
regressor = best_rf
regressor.fit(X_train, Y_train)

# predict values
Y_pred = regressor.predict(X_test)
Y_test.values[:5].swapaxes(0, 1), Y_pred[:5]
```

```python
# inspect example estimator (tree)
import math
trees = regressor.estimators_
print(f"num_trees={len(trees)}")

node_counts = [tree.tree_.node_count for tree in trees]
depths = [tree.get_depth() for tree in trees]
leaves_counts = [tree.get_n_leaves() for tree in trees]

avg_nodes = math.fsum(node_counts) / len(trees)
avg_depth = math.fsum(depths) / len(trees)
avg_leaves = math.fsum(leaves_counts) / len(trees)

print(f"{avg_nodes=}")
print(f"{avg_depth=}")
print(f"{avg_leaves=}")
```

```python
# estimate tuned RF using RSME (normalized)
from sklearn.metrics import mean_squared_error

rsme = np.sqrt(mean_squared_error(Y_test, Y_pred))
rsme_norm = rsme / _range

print(f"{rsme=}")
print(f"{rsme_norm=}")
```
