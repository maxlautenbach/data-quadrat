```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
import data_preprocessing
energy_consumption_data_train, energy_consumption_data_test = data_preprocessing.preprocess_data("Energy Consumption Dataset.csv")
```

```python
energy_consumption_data_train.head()
```

```python
energy_consumption_data_test.head()
```

```python
# Split data into features and target. We'll consider all features apart from "Energy Consumption"

# for training data
X_train = pd.DataFrame(energy_consumption_data_train.iloc[:, :18])
Y_train = pd.DataFrame(energy_consumption_data_train.iloc[:, -1])

# for testing data
X_test = pd.DataFrame(energy_consumption_data_test.iloc[:, :18])
Y_test = pd.DataFrame(energy_consumption_data_test.iloc[:, -1])
```

```python
# Compute range of target variable. Will be used later to normalize RSME
_range = Y_test.max() - Y_test.min()
_range
```

```python
# create estimator
from sklearn.linear_model import LinearRegression
estimator = LinearRegression()
```

```python
# train estimator
estimator.fit(X_train, Y_train)

# predict values
Y_pred = estimator.predict(X_test)
```

```python
# evaluate example using (normalized) RMSE
from numpy import sqrt
from sklearn.metrics import mean_squared_error

rsme = sqrt(mean_squared_error(Y_test, Y_pred))
rsme_norm = rsme / _range
```
